### Model Architecture and Training Strategy

#### 1. Solution Design Approach

The overall strategy for deriving a model architecture was to test different model from simplest one (such as one layer neural network) to more sophisticated model(NVIDIA Network).

My first step was to use a convolution neural network model similar to the LeNet I thought this model might be appropriate because it works well to recognize the German Traffic Sign from previous project. Then I tried more complex architecture derived from NVIDIA Network which is specifically used for the steering angle prediciton of self-driving vehicle. It works much better than LeNet since it has deeper network so that it can capture more detailed information from the images.

In order to gauge how well the model was working, I split my image and steering angle data into a training and validation set. I found that my NIVIDIA model had a low mean squared error on the training set but a relatively high mean squared error on the validation set. This implied that the model was overfitting. 

To combat the overfitting, I collected more data by using images from three cameras, which reduce the overfitting a lot.

The final step was to run the simulator to see how well the car was driving around track one. There were a few spots where the vehicle fell off the track and there is too much jitter in steering. In order to improve the driving behavior in these cases, I added a low-pass filter for the steering data to get rid of the steering angles which is too big affecting the smoothness of the driving.

At the end of the process, the vehicle is able to drive autonomously around the track without leaving the road and it can go back to the center of the road when it is approaching to the side of the road.
